```python
import folium
import pandas as pd
```

```python
## Country coordinates for plotting
country_geo = 'world_countries.json'
```

```python
# Read in the World Development Indicators Database
data = pd.read_csv('DV_num_of_entries.csv')
data.shape
```

```python
data.head()
```

```python
country_codes = pd.read_csv('Country_codes.csv')
country_codes.shape
```

```python
data_1 = data[['Country','average']]
data_1.head()
```

```python
data_to_plot = data_1.merge(country_codes, how = 'left', on = 'Country')
data_to_plot = data_to_plot[['average','CountryCode']]
with pd.option_context("display.max_rows", 1000):
    display(data_to_plot)
```

```python
data_to_plot['average'].value_counts(bins = 10)
```

```python

map = folium.Map(location=[100, 0], zoom_start=1.5)
```

```python
map.choropleth(geo_data=country_geo, data=data_to_plot,
             columns=['CountryCode', 'average'],
             key_on='feature.id',
             fill_color='PuBuGn', fill_opacity=0.7, line_opacity=0.2,
             
            )
```
