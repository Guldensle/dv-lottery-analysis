```python
import pandas as pd
ent_16_18 = pd.read_csv('data/16_18.csv', header = None)
ent_16_18 = ent_16_18.drop(ent_16_18.columns[[1,2,4,5,7,8]], axis =1)
ent_16_18.columns = ['Country', '2016', '2017', '2018']

ent_13_15 = pd.read_csv('data/13_15.csv', header = None)
ent_13_15 = ent_13_15.drop(ent_13_15.columns[[1,2,4,5,7,8]], axis =1)
ent_13_15 = ent_13_15.drop([0,1])
ent_13_15.columns = ['Country', '2013', '2014', '2015']

ent_10_12 = pd.read_csv('data/10_12.csv', header = None)
ent_10_12 = ent_10_12.drop(ent_10_12.columns[[1,2,4,5,7,8]], axis =1)
ent_10_12 = ent_10_12.drop([0,1])
ent_10_12.columns = ['Country', '2010', '2011', '2012']
```

```python
combined_1 = ent_10_12.merge(ent_13_15, how = 'left', on = 'Country')
combined = combined_1.merge(ent_16_18, how = 'left', on = 'Country')
combined.head()
```

```python
combined = combined[combined.Country != 'Foreign State of Chargeability']
combined = combined[combined.Country.notna()]
combined = combined.reset_index(drop=True)
```

```python
# Normalize text
import re

# Replace non-word characters with empty space
def normalize_text(text):
    text = str(text)
    text = re.sub('[^A-Za-z0-9\s]', ' ', text)
    return text

# Text to title format
def title_text(text):
    text = text.title()
    text = text.rstrip()
    return text

# Text to integer
def normalize_value(text):
    text = re.sub('\s', '', text)
    try:
        value = int(text)
    except:
        value = 0
    return value
```

```python
# Apply the functions to the dataframe
combined.iloc[:, 1:] = combined.iloc[:, 1:].applymap(normalize_text)
combined.iloc[:, 1:] = combined.iloc[:, 1:].applymap(normalize_value)

# Apply the function to 'Country' column
combined.iloc[:, 0] = combined.iloc[:, 0].apply(title_text)
```

```python
combined
```

```python
# Calculate number of applications difference between 2010 and 2018
combined['8_years_diff'] = combined['2018'] - combined['2010']

# Calculate the percentage of increase/decrease from 2010 to 2018 in the number of applications 
combined['%_increase'] = (combined[combined['2018'] > 0]['2018'] / combined[combined['2010'] > 0]['2010'] *100) - 100

# Replace null values with 0
combined = combined.fillna(0)

# Calculate the average number of applications for every country
combined['average'] = round(combined.iloc[:,1:10].mean(axis = 1))
combined = combined.sort_values(['average'], ascending = False)

# Rename the dataframe
DV_num_of_entries = combined
```
